# Generate synthetic time series data for benchmarking visualization tools

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Configuration
width_px = 600
height_px = 150
dpi = 100
figsize = (width_px / dpi, height_px / dpi)

# 2. Data Generation
# np.random.seed(42)
x = np.linspace(0, 10, 600)
# Simulate a temperature-like waveform
y = np.sin(x) * 0.5 + np.sin(2.5 * x) * 0.3 + np.random.normal(0, 0.1, size=len(x))

# 3. Plotting
fig, ax = plt.subplots(figsize=figsize, dpi=dpi)

# Change color: use a lighter blue '#81D4FA' (original was '#4A90E2')
line_color = '#81D4FA'

ax.plot(x, y, color=line_color, linewidth=2)
# Fill the area using the same light blue with low opacity
ax.fill_between(x, y, alpha=0.1, color=line_color)

# 4. Styling (transparent background, no axes)
ax.axis('off')
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
fig.patch.set_alpha(0.0) # Make the entire figure background transparent

# 5. Save
output_filename = 'time_series_waveform.svg'
plt.savefig(output_filename, transparent=True, dpi=dpi)
plt.show()
```
