```python
from rich.progress import Progress, BarColumn, TextColumn
from time import sleep
from IPython.display import display, clear_output

stations = range(10)

progress = Progress(
    TextColumn("[white]{task.description}"),
    BarColumn(bar_width=None, complete_style="white on blue", finished_style="white on blue"),
)

display(progress)

task = progress.add_task("Processing Stations", total=len(stations))

for s in stations:
    sleep(0.2)
    progress.update(task, advance=1)
    progress.refresh()
```

```python
from tqdm.rich import tqdm
import time

stations = range(20)

for s in tqdm(
    stations,
    desc="Processing Stations",
    total=len(stations),
    # colour="#90ee90"
):
    time.sleep(0.1)
```

```python
from tqdm.notebook import tqdm
import time
import sys

stations = range(20)

for s in tqdm(
    stations,
    desc="Processing Stations",
    total=len(stations),
    colour="lightblue",
    file=sys.stdout,
    # colour="#90ee90"  # 浅绿色 HEX
):
    time.sleep(0.1)
```
