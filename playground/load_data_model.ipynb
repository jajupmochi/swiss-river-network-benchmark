```python
from pathlib import Path
CUR_ABS_DIR = Path.cwd().parent.absolute()
print(CUR_ABS_DIR)
```

```python
# Read a graph from a file:
import torch

def read_graph(graph_name):
    # return torch.load(f'swissrivernetwork/journal/dump/graph_{graph_name}.pth')
    return torch.load(CUR_ABS_DIR / f'../datasets/graph_{graph_name}.pth')

data, edges = read_graph('swiss-2010')  # swiss-1990, swiss-2000, zurich
print(data)
print(edges)
```

```python
# Read a csv data:
import pandas as pd

def read_csv_train(graph_name):
    # return pd.read_csv(f'swissrivernetwork/journal/dump/{graph_name}_train.csv')
    return pd.read_csv(CUR_ABS_DIR / f'../datasets/{graph_name}_train.csv')

df = read_csv_train('swiss-1990')
print(df)
```
